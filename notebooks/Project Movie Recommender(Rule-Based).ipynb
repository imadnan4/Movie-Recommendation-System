```python
movies = movies[['id','title','overview','genres','keywords','tagline', 'production_companies']]
```

```python
movies.head()
```

```python
def remove_spaces_from_column(df, column_name):
    df[column_name] = df[column_name].str.replace(" ", "", regex=False)
    return df
```

```python
movies = remove_spaces_from_column(movies, 'genres')
movies = remove_spaces_from_column(movies, 'production_companies')
```

```python
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
movies.sample(5)
```

```python
movies.head()
```

```python
# Ensure all columns used for concatenation are of type string
#movies['overview'] = movies['overview'].apply(lambda x:x.split())
movies['tags'] = (
    movies['overview'].astype(str) + " " +
    movies['keywords'].astype(str) + " " +
    movies['tagline'].astype(str) + " " +
    movies['genres'].astype(str) + " " +
    movies['production_companies'].astype(str)
)

# Print sample of the resulting 'tags' column
print(movies['tags'].head())


movies = movies.apply(lambda x: x.replace(",", ", ") if isinstance(x, str) else x)
```

```python
movies.head()
```

```python
new = movies.drop(columns=['overview','genres','keywords','tagline','production_companies'])
```

```python
new.head()
```

```python
new.shape
```

```python
new['tags'] = new['tags'].str.strip('[]').str.replace("'", "").str.replace(",", "")
new.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(new['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
def recommend(movie):
    index = new[new['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
